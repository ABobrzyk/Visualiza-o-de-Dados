▶ **Gráfico de Dispersão**

É uma representação visual que mostra a relação entre duas variáveis. Geralmente, exibe pontos no plano cartesiano, sendo que cada ponto representa uma observação com valores para as duas variáveis. Se usarmos também forma, tamanho ou cor é possível representar mais de dois atributos.

No *Plotly Express* usamos a função *scatter* para criar um gráfico de dispersão, conforme mostra o exemplo abaixo.

Informações deste tipo de gráfico estão disponíveis em:
*   https://plotly.com/python/line-and-scatter/
*   https://plotly.com/python-api-reference/generated/plotly.express.scatter.html

````python
fig = px.scatter(df, x='sepal_length', y='sepal_width', size='petal_length',
                 title='Gráfico de dispersão',
                 labels={'sepal_length': 'Sepal Length', 'sepal_width': 'Sepal Width'},
                 color='species',
                 symbol='petal_width')
fig.show()
````

▶ **Matriz de Dispersão**

Esta representação visual é usada para exibir múltiplos gráficos de dispersão em uma matriz, sendo que cada um mostra a relação entre duas variáveis de um conjunto de dados multivariado (ou multidimensional).

No *Plotly Express* usamos a função *scatter_matrix* para criar uma matriz de dispersão, conforme mostra o exemplo abaixo.

Informações deste tipo de gráfico estão disponíveis em:
*   https://plotly.com/python/splom/
*   https://plotly.com/python-api-reference/generated/plotly.express.scatter_matrix.html

````python
fig = px.scatter_matrix(df, title='Matriz de dispersão', height=800)
fig.show()
````

▶ **Gráfico de Pizza**

Este tipo é usado para representar proporções ou porcentagens de diferentes categorias em relação ao todo.

No exemplo abaixo, é apresentado o percentual de cada espécie do DataFrame.

Informações deste tipo de gráfico estão disponíveis em:
*   https://plotly.com/python/pie-charts/
*   https://plotly.com/python-api-reference/generated/plotly.express.pie.html

````python
# Calcula o percentual de cada espécie
percentuais = df['species'].value_counts(normalize=True) * 100

# Cria um gráfico de pizza com os percentuais das espécies
fig = px.pie(names=percentuais.index, values=percentuais.values, title="Percentual de cada espécie")
fig.show()
````

▶ **Gráfico de Radar**

Este tipo de representação visual exibe várias variáveis em um eixo circular, mostrando como cada variável contribui para um determinado registro.

O código abaixo ilustra como criar um gráfico de radar para um registro.

Informações deste tipo de gráfico estão disponíveis em:
*   https://plotly.com/python/radar-chart/
*   https://plotly.com/python-api-reference/generated/plotly.express.line_polar.html

````python
# Seleciona um registro específico (por exemplo, o primeiro registro)
registro_selecionado = df.iloc[1]

# Extrai as medidas (comprimentos de sépala e pétala) do registro selecionado
medidas = registro_selecionado[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]

# Cria um DataFrame para o registro selecionado
df_registro = pd.DataFrame({'variavel': medidas.index, 'valor': medidas.values})

# Cria um gráfico de linhas polar com as coordenadas polares para o registro selecionado
fig = px.line_polar(df_registro, r='valor', theta='variavel', line_close=True,
                    title="Gráfico de Radar para um Registro")

# Preenche o polígono formado pelos valores
fig.update_traces(fill='toself')
fig.show()
````
